```python
import os
from glob import glob
```

```python
from transformers import pipeline

# Load a pre-trained model and tokenizer
qa_pipeline = pipeline("question-answering", model="bert-large-uncased-whole-word-masking-finetuned-squad")
```

inference for example

```python
# Example data
context = "The hotel is located just off the main square and features free Wi-Fi, a rooftop pool, and a spa."
question = "What amenities does the hotel offer?"

# Get the answer from the model
result = qa_pipeline(question=question, context=context)
print("Answer:", result['answer'])
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("allenai/llama")
model = AutoModelForCausalLM.from_pretrained("allenai/llama")

# Example query
context = "The hotel is located near the beach and offers free breakfast."
question = "What amenities does the hotel offer?"
input_text = f"Context: {context}\nQuestion: {question}"

# Encode and generate response
inputs = tokenizer.encode(input_text, return_tensors='pt')
outputs = model.generate(inputs, max_length=150)
response = tokenizer.decode(outputs[0], skip_special_tokens=True)

print(response)
```
